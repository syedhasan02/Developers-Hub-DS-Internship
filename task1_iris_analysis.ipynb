# Task 1: Exploring and Visualizing Iris Dataset

## Objective
The objective of this task is to understand how to load, explore, and visualize a dataset using pandas, matplotlib, and seaborn.

## Dataset
The Iris dataset contains measurements of flowers such as sepal length, sepal width, petal length, and species.

```python
#                                    Task 1: Exploring and Visualizing a Simple Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset('iris')
# iris = pd.read_csv('iris.csv') If teacher wants CSV method:

print(iris.shape) # Shape (rows, columns)
print(iris.columns) # Column Names
print(iris.head()) # First 5 Rows

print(iris.describe()) # This shows: Mean, Min, Max, Standard deviation

# Scatter Plot(Relationship between variables) --> Sepal Length vs Petal Length
sns.scatterplot(x='sepal_length', y='petal_length', data=iris)
plt.title("Sepal Length vs Petal Length")
plt.show()

# Histogram shows how values are distributed.
plt.figure()
sns.histplot(iris['sepal_length'])
plt.title("Distribution of Sepal Length")
plt.show()

# Box plot shows median, quartiles, and outliers.
plt.figure()
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title("Sepal Length by Species")
plt.show()

# Color-based scatter plot ; This separates species with colors
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris)
plt.show()

```

## Dataset Understanding
We explore the structure of the dataset using shape, columns, and first few rows.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

X = iris.drop('species', axis=1)
y = iris['species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

## Conclusion

- The dataset was successfully loaded and analyzed.
- Scatter plot showed relationships between variables.
- Histogram showed distribution of values.
- Box plot helped identify spread and outliers.
- The model achieved good accuracy in classifying species.
